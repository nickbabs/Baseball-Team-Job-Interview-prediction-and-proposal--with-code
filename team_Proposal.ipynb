## Introduction
In this proposal, I present an analytical approach on how one could answer the following question provided by the infield coach: 
<p>“One of our infielders, Player X, seems to be struggling in the field. He’s got a great arm, but he’s made a few errors this season and is failing to get to some balls. Could you look into this and identify any problem areas that we can target with drills?”

#### Proposed Approaches

Player X has proven to have a strong arm, but he seems to be struggling to locate the ball off the bat and has been responsible for a few errors this season. We will look at each play where Player X committed an error as well as the plays he seemed to be slow getting to the ball. It is imperative to determine each play's level of difficulty in order to analyze why it was a more difficult play for him to make. We will then attempt to identify potential problem areas for his fielding struggles in order to determine the drills necessary to aid his improvement. This will be done by examining the following:

1. We shall examine the effect of the batter by noting what side of the plate he swings from, the exit velocity and launch angle of the batted ball, and the scouting report on the batter (average exit velocity, average launch angle, batting average, etc.). This will give us an idea if Player X is struggling more against rightys or leftys. Looking at the exit velocity and launch angle of the batted balls will tell us if Player X is struggling against more hard hit balls or soft hit balls, as well as if he is struggling more against line drives or ground balls. The scouting report influences the expectations of the fielders and if there's an unexpected outcome, the player could be caught off guard, causing a delay in his reaction time. 

2. When examining the effect of the pitcher, it's important to note what pitch was thrown, the speed of the pitch, and the actual location of the pitch relative to its intended location. This will give us an idea of how Player X is seeing the ball when the pitcher throws certain pitches and how different pitches effect his reaction time. We also need to look at the location of the pitch as that could have a direct impact on the direction of the batted ball. If the catcher sets up in one spot and the pitcher mistakenly throws it to another, the fielder could be favoring the side he expects the ball to go to which could make the play more difficult if the ball is batted elsewhere.

3. We can look at the infield position Player X was at and what sort of shift, if any, was implemented on that particular play. If the player is not in his natural position, he may simply need more experience at that position. The location of the batted ball relative to his body is also worth taking a look at. If the missed ball was hit on his left or right side, it would be easy to replicate this in a drill. If the player failed an attempt to backhand the ball, he may need more practice with that. We should also determine if the player was running or standing still while fielding the ball. As previously mentioned, the exit velocity and launch angle of the ball on this play is also imperative and must be considered while analyzing these mentioned situations.

4. Examining different game situations such as how many men were on base, what inning the game was in, the score of game, and the implied pressure of the moment is another way to determine the difficulty of the play. The game of baseball is just as much mental as it is physical, and if Player X is determined to not be able to handle certain moments or gameplay situations, these situations can be replicated in a drill.

5. Another important detail to be examined is what type of grass is on the infield of the stadium. The ball tends to bounce differently on different surfaces and Player X may need more repition fielding the ball off of certain field conditions.

6. Examine the player's workload leading up to the game. Taking a look at how many games the player has played could give us an idea of his body's fatigue. Is the player not getting enough rest and thus impacting the level of his play? Does the player need better conditioning?

7. Examine the time of game, position of the sun or the stadium lighting, and the game's weather conditions. If the player's performance seems to decline in certain environments, the team can attempt to replicate these conditions to give Player X more repetition in the conditions where he is struggling.

The approaches listed above would be most effective if they were all taken into consideration as opposed to looking at just one. Each approach was detailed in a manor assuming that we were looking at the plays where Player X committed an error or where he seemed to struggle to make the play. We need to focus on the player's vision of the ball and his ability to react quickly to each batted ball. We also touched upon the mental aspect of the game and if the given game situation negatively impacts Player X's ability to perform. 
<p> Each approach listed above should be easily replicated in a drill or simulation scrimmage, allowing for Player X to receive the attention he needs in the problem areas that have been determined.

#### Issues

There is always the potential of certain issues when recording and analyzing data, and sometimes assumptions must be made that are not very easy to determine.
- The first thing to think about is the commonality of human error and basic human behavior. As I mentioned earlier, there is also a mental aspect to this game and it is hard to determine if a player is having an off day due to his mind not being focused on the game. This is next to impossible to measure and quantify when analyzing our dataset. There is also the potential of the official scorer or whomever is recording this data to make a mistake while doing so.
- Another potential issue could be determining if a batted ball was a play that Player X was capable of making. While it is easy to measure when the official scorer ruled a play as an error, it is not as easy to measure a play as one that Player X was supposed to be able to make. This will provide some difficulty and possibly even some outliers when determining which plays are to be recorded as part of our dataset.
- There could be some challenge in determining whether a pitcher missed his spot or not because he may not tell the truth when being questioned, but it should be easy to watch game film and see where the catcher sets up his glove and compare it to where he actually catches the ball. This again can provide difficulty, because some pitchers may like the catcher to set up in one position when they are actually hoping for the ball to end up in another.
- A player's level of fatigue won't always amount to how many games they've played without having an off day. Many other factors can effect fatigue such as a player's diet or things that go on in their personal life outside of baseball. This makes fatigue difficult to measure and determine as an underlying factor of an error being made or a missed play.
- A player may not always be honest if the sun or the stadium lights were a reason for an error or a missed play, making this feature a difficult one to determine in not so obvious situations. 

#### Methodology

In a perfect world, each approach proposed above would be possible and there would be no doubts in our dataset. But in reality, a lot of assumptions must be made for this to be true, some of which have no way of measuring if these assumptions are true or not. Because of these reasons, the approaches listed above that I would suggest be used in this analysis would be some combination of 1, 2, 3, and 5.

In order to determine whether or not the plays where an error is not occured are an area of struggle for Player X, we could look at datasets for average fielders within the MLB, at the same position, and compare their results on similar plays to Player X's performance. If Player X's performance seems to be below the average, it can be determined that this is an area of struggle. When determining who is an average fielder or not, we could look at the golden glove votes for each position from the previous year. Since there are 32 teams, we will want to examine the players that ranked in the second and third quartile (9 - 24) in votes at their position.

First we should look at the batter's exit velocity of the batted ball. If the exit velocity was above average, it can be said that the error was due to a hard hit ball. This would then tell the team that Player X needs more practice with hard hit balls. If the exit velocity is average, the next step would be to determine the launch angle of the batted ball. If the Player X is consistently missing batted balls with similar launch angles, we know this is an area for him to get more practice in. Then we can take a look at the type of grass that was played on and look for consistencies there.  If none of the above seem to be a consistent issue, we can then look at the pitch that was thrown and the position the player is in. If Player X is consitently performing poorly on pitches that are in a certain location (up and away, low and inside, etc.) or that are a certain speed, then that would be a good indication that this pitch is a target area for the player to practice. Next we should look at the type of pitches that were thrown on the plays Player X performed poorly on to determine if he is having trouble seeing breaking balls or fastballs. If these pitch features do not have a high correlation to Player X's poor fielding, we could also check his positioning in the field. We want to visualize where the player is standing on the infield so we can split the infield into different sectors, or bins. If the player makes more mistakes when in a certain location, while also taking into consideration if the batter was a righty or lefty, then it can be determined that this area should be worked on. This can be done in a simulation scrimmage or a mock game situation with a batter at the plate who has similar attributes (average exit velocity, average launch angle, etc.) to the batter from the play where Player X performed poorly.

```python
y_pred = predictions2
```

```python
from sklearn import metrics

mae = metrics.mean_absolute_error(y_test, y_pred)
print("Mean Absolute Error:", mae)

# Calculate the Mean Squared Error
mse = metrics.mean_squared_error(y_test, y_pred)
print("Mean Squared Error:", mse)

# Calculate the Root Mean Squared Error
rmse = np.sqrt(mse)
print("Root Mean Squared Error:", rmse)

# Calculate the R-squared value
r2 = metrics.r2_score(y_test, y_pred)
print("R-squared:", r2)
```
